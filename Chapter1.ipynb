# **1.3 머신러닝 시스템의 종류**  

**1.3.1 지도 학습과 비지도 학습**  
1) 지도 학습 : 훈련데이터에 레이블(원하는 답) 포함  
2) 비지도 학습 : 훈련데이퍼에 레이블 없음  
3) 준지도 학습 : 레이블이 없는 데이터 >> 레이블이 있는 데이터(아주 조금)  
4) 강화 학습 : 보상과 벌점이 있는 시스템(에이전트)  

**1.3.2 배치 학습과 온라인 학습**  
 1) 배치 학습(오프라인 학습) : 점진적 학습 불가, 가용한 모든 데이터 사용해 훈련, 오프라인에서 수행  
 2) 온라인 학습 : 데이터를 순차적으로 한개씩(or 미니배치 = 작은 묶음)단위로 훈련, 학습 단계가 빠르고 비용이 적게 듦  

 **1.3.4 사례 기반 학습과 모델 기반 학습**  
 1) 사례 기반 학습 : 시스템이 사례를 기억함, 유사도 측정으로 새로운 데이터 일반화  
 2) 모델 기반 학습 : 샘들플의 모델을 만들어 예측에 사용
